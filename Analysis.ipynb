```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('output_times.csv')
```

```python
df
```

```python
df_256 = df[df['index_records_per_node'] == 256]
plt.plot(list(df_256.num_threads), list(df_256.average_time_elapsed))
plt.xlabel('# Threads')
plt.ylabel('Avg. Time (seconds)')
plt.title('Times for 256 index records per node and 3 levels')
```
